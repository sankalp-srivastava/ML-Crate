### Finding Top 15 Languages Based on Number of Repos

```python
top_15 = repos_df.sort_values(by="num_repos", ascending=False)[:15]
top_15
```

### We can see that JavaScript, CSS and HTML are the top 3 languages when it comes to number of repos, followed by Shell and Python

```python
plt.figure(figsize=(15,10))
plt.title("Top 15 Programming languages Repos in Github")
sns.barplot(x=top_15['language'], y=top_15['num_repos'])
```

### Visualizing Number of Issues with Respect to Each Year

```python
total = issues_df.groupby(['year']).sum()/1000
plt.figure(figsize=(10,5))
plt.plot(issues_df['year'],issues_df['count'])
plt.title("Number of Issues With Respect to Year")
plt.xlabel('Year')
plt.ylabel('Issues')
```

We can see that during 2015-2016, there are more number of issues.

### Visualizing Total Number of PRs with respect to each Year

```python
total = prs_df.groupby(['year']).sum()/1000
plt.figure(figsize=(10,5))
plt.plot(prs_df['year'],prs_df['count'])
plt.title("Number of PRs  With Respect to Year")
plt.xlabel('Year')
plt.ylabel('PRs')
```

We can see that during 2017, there are more number of PRs

### Top 15 Languages which has most number of Issues

```python
top_issues = issues_df.groupby('name')['count'].sum().nlargest(15)
top_issues
```

### Top 15 Languages which has most number of PRs

```python
top_prs = prs_df.groupby('name')['count'].sum().nlargest(15)
top_prs
```

### Issues of TOP 8 Languages in Github

```python
top_8_issues = issues_df.groupby(['name']).sum().sort_values('count', ascending=False)['count'][:8].index
top_8_issues
```

```python
top_8_issues_each = issues_df[issues_df.name.isin(top_8_issues)]
top_8_issues_each
```

```python
plt.figure(figsize=(15,10))
sns.set_theme(style='darkgrid')
ax = sns.lineplot(x='year',y='count',hue='name',data=top_8_issues_each)
ax.axes.set_title(f'Issues of Top 8 Languages in Github',fontsize=15)
ax.set_xlabel("Year",fontsize=12)
ax.set_ylabel("Issues Count",fontsize=12)
```

### PRs of TOP 8 Languages in Github

```python
top_8_prs = prs_df.groupby(['name']).sum().sort_values('count', ascending=False)['count'][:8].index
top_8_prs
```

```python
top_8_prs_each = prs_df[prs_df.name.isin(top_8_prs)]
top_8_prs_each
```

```python
plt.figure(figsize=(15,10))
sns.set_theme(style='darkgrid')
ax = sns.lineplot(x='year',y='count',hue='name',data=top_8_prs_each)
ax.axes.set_title(f'PRs of Top 8 Languages in Github',fontsize=15)
ax.set_xlabel("Year",fontsize=12)
ax.set_ylabel("PRs Count",fontsize=12)
```
