```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df1 = pd.read_csv("gate_applicants.csv")
df1.head()
```

```python
df1.drop(columns=["Unnamed: 0"],inplace=True)
df1.head()
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df1.columns
```

```python
df1.isna().sum(axis=0)
```

```python
df1.corr()
```

```python
plt.figure(figsize=(10, 10))
heatmap = sns.heatmap(df1.corr(), vmin= -1, vmax = 1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12})
```

```python
year = sorted(list(set(df1["Year"])))
fig1,axes1 = plt.subplots(3,figsize=(15,15))
for i in year:
  dt = df1[df1["Year"]==i]
  axes1[year.index(i)].plot(dt["Paper Code"],dt["Applied"])
  axes1[year.index(i)].set_xlabel("Paper Codes")
  axes1[year.index(i)].set_ylabel("Applied")
  axes1[year.index(i)].set_title(i)
```

```python
fig2,axes2 = plt.subplots(3,figsize=(15,15))
for i in year:
  ds = df1[df1["Year"]==i]
  axes2[year.index(i)].plot(ds["Paper Code"],ds["Appeared"])
  axes2[year.index(i)].set_xlabel("Paper Codes")
  axes2[year.index(i)].set_ylabel("Appeared")
  axes2[year.index(i)].set_title(i)
```
