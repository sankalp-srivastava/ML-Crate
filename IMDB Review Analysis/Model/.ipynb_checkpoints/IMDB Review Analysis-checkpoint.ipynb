# Data Preprocesing

```python
# Now Lets remove the Punctuation , stopwords, noisy_texts from the reviews.
def remove_punctuation_mark(text):
    filtered_words = word_tokenize(text)
    for mark in string.punctuation:
        filtered_words = [word.replace(mark,'') if mark in word else word for word in filtered_words] 
    filtered_text = " ".join(filtered_words)
    return filtered_text


imdb_data['review']= imdb_data['review'].apply(remove_punctuation_mark)
```

```python
import nltk
nltk.download('punkt')
```

```python
# we have removed punctutation from the text succesfully.
imdb_data['review'][0]
```

```python
imdb_data['review'] = imdb_data['review'].str.replace(' br ', '')
```

```python
imdb_data['review'][0]
```

```python
# Lets remove stopwords from the reviews
eng_stopwords = set(stopwords.words('english'))
def remove_stopwords(text):
    filtered_words = text.split()
    filtered_words = [word for word in filtered_words if word not in eng_stopwords]
    return " ".join(filtered_words)

imdb_data['review'] = imdb_data['review'].apply(remove_stopwords)
```

```python
import nltk
nltk.download('stopwords')
```

```python
# let see the stopwords are removed or not`
imdb_data['review'][0]   # yes, stopwords is removed. great almost we have clean the data.
```

```python
#Lets make all the characters to the lowercase.
imdb_data['review'] = imdb_data['review'].str.lower()
```

```python
# Lets check the data is converted into the lower case or not 
imdb_data['review'][0]    # yes, great it has been converted.
```

```python
imdb_data.head()    # we have reached to the milestone we want to achieve.
```

```python
# Lets find some HTTP word which exists in the text reviews or not?
count_http_tags=0
index=0
index_store=[]
for review in imdb_data['review']:
    if 'http' in review:
        count_http_tags+=1
        index_store.append(index)
    index+=1
```

```python
print('No of reviews with URL Links:', count_http_tags)
print("Index at which the UrL links appears:\n", index_store)
```

```python
imdb_data['review'][907]  # Can see the url link in all these index.
```

```python
# Lets Remove the Url Links from the data reviews.
def remove_urllinks(text):
    filtered_text = text
    if 'http' in text:
        url_content = re.findall(r'http www\w+',filtered_text)
        for url in url_content:
            filtered_text = filtered_text.replace(url,'')
    return filtered_text

imdb_data['review'] = imdb_data['review'].apply(remove_urllinks)
        
            
```

```python
# Lets convert now words in vectors using word2Vec from Gensim Library
# let Merge every review in one list 
all_review_words = []
for review in imdb_data.review:
    review_words=word_tokenize(review)
    all_review_words.append(review_words)
```

```python
# Now all the reviews has been converted into words and stored in the list
all_review_words[49999]
```

# Word2Vec Model

```python
# Lets make an object of word2vec model with passing arguments of all_review_words with min_count=1.
w2v_model = Word2Vec(all_review_words, min_count=1)
```

```python
len(w2v_model.wv.index_to_key)   # total these many words are there in the review column which has been converted into vectors.
```

```python
# Lets make the average of words of list so that we can make one single vector representation for each an every review
def compute_avg_vector_repr(all_review_words):
    avg_vector_of_each_review =[]
    for review_words in all_review_words:
        vector_of_words = w2v_model.wv[review_words]
        avg_vector_of_each_review.append(vector_of_words.mean())
    
    return avg_vector_of_each_review

avg_vector_repr = compute_avg_vector_repr(all_review_words)
imdb_data['review_vectors'] = avg_vector_repr
```

# Splitting the data for Feature and target variables.

```python
imdb_data_vectors.isnull().sum() 
# while imputing avg values there was no missing values occured it means that every value has been imputed properly
```

```python
feature = np.array(imdb_data_vectors['review_vectors']).reshape(-1,1)
target = np.array(imdb_data_vectors['sentiment']).reshape(-1,1)
```

```python
feature.shape, target.shape
```

```python
# lets apply train_test_split from sklearn 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42)
print(x_train.shape, y_train.shape, x_test.shape,y_test.shape)
```

# Applying SVM classifier.

```python
from sklearn.svm import SVC

svc = SVC()    # applying default kernel i.e rbf( radial basis function )
svc.fit(x_train, y_train)
```

```python
# Predictions on the test data.
y_pred = svc.predict(x_test)
```

## Model Evaluation for kernel RBF

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def metrics(y_test, y_pred):
    print('Accuracy Score:', accuracy_score(y_test, y_pred))
    print('Classification Report:\n',classification_report(y_test, y_pred))
    print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
```

```python
metrics(y_test, y_pred)
```

### 60.22% accuracy has attained when we have apply svm using default kernel called 'rbf'.

```python
# Lets change the kernel now:
svc1 = SVC(kernel ='linear') #lets fit the model by using the linear kernel
svc1.fit(x_train, y_train)

y_pred1 = svc1.predict(x_test)

#metrics
metrics(y_test,y_pred1)
```

### while using kernel as linear the accuracy has increased to 0.02% which is almost negligible. Accuracy attained by this is 60.24%.

```python
# Lets change the kernel now:
svc2 = SVC(kernel ='sigmoid') #lets fit the model by using the linear kernel
svc2.fit(x_train, y_train)

y_pred2 = svc2.predict(x_test)

#metrics
metrics(y_test,y_pred2)
```

```python
# Lets change the kernel now:
svc3 = SVC(kernel ='poly') #lets fit the model by using the poly kernel
svc3.fit(x_train, y_train)

y_pred3 = svc3.predict(x_test)

#metrics
metrics(y_test,y_pred3)
```

### Accuracy scored when we used this with kernel poly, the score we have attained is 54.76%
